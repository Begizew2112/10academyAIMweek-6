```python

# 1. Convert 'TransactionStartTime' to datetime format
df['TransactionStartTime'] = pd.to_datetime(df['TransactionStartTime'], format='%Y-%m-%dT%H:%M:%SZ')

# 2. Now extract the components (hour, day, month, year)
df['TransactionHour'] = df['TransactionStartTime'].dt.hour
df['TransactionDay'] = df['TransactionStartTime'].dt.day
df['TransactionMonth'] = df['TransactionStartTime'].dt.month
df['TransactionYear'] = df['TransactionStartTime'].dt.year

# Display the extracted features
df[['TransactionHour', 'TransactionDay', 'TransactionMonth', 'TransactionYear']].head()
```

perform encoding for the columns ProductId

```python
# Perform one-hot encoding for 'ProductId'
product_encoded = pd.get_dummies(df['ProductId'], prefix='ProductId')

# Merge back and drop the original column
df_encoded = pd.concat([df, product_encoded], axis=1)

```

productid encoding

```python
# Convert True/False to 0/1 for the encoded columns
encoded_columns = [ 'ProductId_ProductId_1', 'ProductId_ProductId_10',
       'ProductId_ProductId_11', 'ProductId_ProductId_12',
       'ProductId_ProductId_13', 'ProductId_ProductId_14',
       'ProductId_ProductId_15', 'ProductId_ProductId_16',
       'ProductId_ProductId_19', 'ProductId_ProductId_2',
       'ProductId_ProductId_20', 'ProductId_ProductId_21',
       'ProductId_ProductId_24', 'ProductId_ProductId_3',
       'ProductId_ProductId_4', 'ProductId_ProductId_6',
       'ProductId_ProductId_9']

df_encoded[encoded_columns] = df_encoded[encoded_columns].astype(int)
```

```python
df_encoded.head()
```

Encoding ChannelId

```python
# Perform one-hot encoding on 'ChannelId' column (assuming it's categorical)
channel_encoded = pd.get_dummies(df['ChannelId'], prefix='ChannelId')

# Merge the encoded columns back into the original DataFrame
df2 = pd.concat([df_encoded, channel_encoded], axis=1)

# Drop the original 'ChannelId' column
#df_encoded.drop('ChannelId', axis=1, inplace=True)
```

```python
df2.head()
```

Convert True/False to 1/0

```python
# Specify the columns to convert from True/False to 0/1
encoded_columns = ['ChannelId_ChannelId_1', 'ChannelId_ChannelId_2', 
                   'ChannelId_ChannelId_3', 'ChannelId_ChannelId_5']

# Convert the specified columns from boolean (True/False) to integers (0/1)
df2[encoded_columns] = df2[encoded_columns].astype(int)
```

```python
df2.head()
```

ProviderId Encoding:

```python
# Perform one-hot encoding for 'ProviderId'
provider_encoded = pd.get_dummies(df['ProviderId'], prefix='ProviderId')

# Merge back into the main DataFrame and drop the original column
df3 = pd.concat([df2, provider_encoded], axis=1)
#df_encoded.drop('ProviderId', axis=1, inplace=True)

# Convert True/False to 0/1 for the encoded columns
encoded_columns = ['ProviderId_ProviderId_1', 'ProviderId_ProviderId_2', 
                   'ProviderId_ProviderId_3', 'ProviderId_ProviderId_4', 
                   'ProviderId_ProviderId_5', 'ProviderId_ProviderId_6']

df3[encoded_columns] = df3[encoded_columns].astype(int)
```

```python
df3.head()
```

PricingStrategy Encoding:

```python
# Perform one-hot encoding for 'PricingStrategy'
pricing_encoded = pd.get_dummies(df['PricingStrategy'], prefix='PricingStrategy')

# Merge back and drop the original column
df4 = pd.concat([df3, pricing_encoded], axis=1)
#df3.drop('PricingStrategy', axis=1, inplace=True)

# Convert True/False to 0/1 for the encoded columns
encoded_columns = ['PricingStrategy_0', 'PricingStrategy_1', 'PricingStrategy_2', 'PricingStrategy_4']

df4[encoded_columns] = df4[encoded_columns].astype(int)
```

```python
df4.head()
```

ProductCategory Encoding:

```python
# Perform one-hot encoding for 'ProductCategory'
product_category_encoded = pd.get_dummies(df['ProductCategory'], prefix='ProductCategory')

# Merge back and drop the original column
df5 = pd.concat([df4, product_category_encoded], axis=1)
#df_encoded.drop('ProductCategory', axis=1, inplace=True)

```

```python
df5.head()
```

```python
# Convert True/False to 0/1 for the encoded columns
encoded_columns = ['ProductCategory_airtime', 'ProductCategory_data_bundles', 
                   'ProductCategory_financial_services', 'ProductCategory_movies', 
                   'ProductCategory_other', 'ProductCategory_ticket', 
                   'ProductCategory_tv', 
                   'ProductCategory_utility_bill']

df5[encoded_columns] = df5[encoded_columns].astype(int)
```

```python
df5.head()
```

```python
# Frequency encode the 'CustomerId' column
customer_id_freq = df5['CustomerId'].value_counts()

# Map the frequencies back to the original dataframe
df5['CustomerId_Encoded'] = df5['CustomerId'].map(customer_id_freq)


```

```python
# Frequency encode the 'CustomerId' column
customer_id_freq = df5['SubscriptionId'].value_counts()

# Map the frequencies back to the original dataframe
df5['SubscriptionId_Encoded'] = df5['SubscriptionId'].map(customer_id_freq)
```

```python
# Frequency encode the 'CustomerId' column
customer_id_freq = df5['AccountId'].value_counts()

# Map the frequencies back to the original dataframe
df5['AccountId_Encoded'] = df5['AccountId'].map(customer_id_freq)
```

```python
# Frequency encode the 'CustomerId' column
customer_id_freq = df5['BatchId'].value_counts()

# Map the frequencies back to the original dataframe
df5['BatchId_Encoded'] = df5['BatchId'].map(customer_id_freq)
```

```python
df5.head()
```

```python
# List of columns to drop (unnecessary for normalization)
columns_to_drop = ['TransactionId', 'BatchId', 'AccountId', 'SubscriptionId', 
                   'CustomerId', 'CurrencyCode', 'CountryCode', 'ProviderId', 
                   'ProductId', 'ProductCategory', 'ChannelId']
                  

# Dropping the columns
df_cleaned = df5.drop(columns=columns_to_drop)

```
